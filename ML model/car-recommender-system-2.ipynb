```python
# to set label for price
def get_label_for_price(price):
    if price < 8:
        return 'EconomicalRange'
    elif 8 < price < 22:
        return 'MidRange'
    elif 22 < price < 26:
        return 'HighRange'
    elif 26 < price < 30 :
        return 'VeryHighRange'
    elif 30 < price < 40:
        return 'Luxry'
    elif 40 < price < 80:
        return 'HighLuxry'
    else:
        return 'UltimateLuxry'
```

# K-Means

```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

dataset = cars_data
```

```python
cars_data.head(1)
```

```python
rows=cars_data['ID'].shape[0]
rows
```

```python

cols=6
A = np.zeros((rows,cols))
cars_data.columns
```

```python
cars_data['row_num'] = np.arange(len(cars_data))
# Remove two columns name is 'C' and 'D'
cars_data=cars_data.drop(['name', 'engine_cc','transmission','doors','brand_name','model_name','year','submodel','photo','segment','description','car_name'], axis = 1)
cars_data.head(10)


df1=cars_data.loc[cars_data['ID']==200700162]
cars_data
```

```python
ID = cars_data.ID.unique().shape[0]

A = np.zeros((ID,5))
for line in cars_data.itertuples():
    A[line[7],0] = line[2]
    A[line[7],1] = line[3]
    A[line[7],2] = line[4]
    A[line[7],3] = line[5]
    A[line[7],4] = line[6]
 
print("Original rating matrix : ",A)
```

```python
avg_highway_milege = cars_data['highway_milege'].mean()
avg_highway_milege
```

```python
avg_city_milege = cars_data['city_milege'].mean()
avg_city_milege
```

```python
avg_msrp = cars_data['msrp'].mean()
avg_msrp
```

```python
avg_rating = cars_data['rating'].mean()
avg_rating
```

```python
avg_consumer_rating = cars_data['consumer_rating'].mean()
avg_consumer_rating
```

```python
for i in range(len(A)):
    for j in range(len(A[0])):
        if j==0:
            avg=avg_highway_milege
        elif j==1:
            avg=avg_city_milege
        elif j==2:
            avg=avg_msrp
        elif j==3:
            avg=avg_rating
        else:
            avg=avg_consumer_rating
        
        if A[i][j]>=avg:
              A[i][j]=1
        else:
              A[i][j]=0
                
                
print(A)
```

```python
csr_sample = csr_matrix(A)
print(csr_sample)
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=5, n_jobs=-1)
knn.fit(csr_sample)
```

```python
dataset_sort_des = cars_data.sort_values(['rating','city_milege'], ascending=[True,True])
filter1 = dataset_sort_des.loc[(dataset_sort_des['rating'] == 5) & (dataset_sort_des['city_milege'] == 30)].row_num
filter1 = filter1.tolist()
```

```python
distances1=[]
indices1=[]
for i in filter1:
    distances , indices = knn.kneighbors(csr_sample[i],n_neighbors=5)
    indices = indices.flatten()
    indices= indices[1:]
    indices1.extend(indices)
print("Items to be recommended: ",indices1)
```

```python
df1=cars_data[cars_data['row_num'].isin(indices1)]
recom_list=df1['ID'].values.tolist()

for val in recom_list:
    df2=dataset.loc[dataset['ID']==val]
    print(df2['car_name'].values[0])
    
```
